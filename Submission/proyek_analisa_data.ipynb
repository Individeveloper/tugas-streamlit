```python
df = pd.read_csv("https://raw.githubusercontent.com/marceloreis/HTI/refs/heads/master/PRSA_Data_20130301-20170228/PRSA_Data_Aotizhongxin_20130301-20170228.csv")
df.head()
```

## Visualization & Explanatory Analysis

### Pertanyaan 1: Berapakah rata-rata suhu selama 5 tahun terakhir?


```python

plt.figure(figsize=(10, 5))
plt.bar(
    df["year"],
    df["TEMP"],
    color="#72BCD4",
    width=0.5
)
plt.title("Suhu Aotizhongxin (2013-2017)", loc="center", fontsize=20)
plt.xlabel("Tahun", fontsize=12)
plt.ylabel("Temperatur", fontsize=12)
plt.xticks(df["year"], fontsize=10)
plt.yticks(fontsize=10)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

### Pertanyaan 2: Berapakah rata-rata curah hujan selama 5 tahun terakhir?

```python
plt.figure(figsize=(10, 5))
plt.plot(
    df["year"],
    df["RAIN"],
    color="#72BCD4",
    width=0.5
)
plt.title("Curah Hujan Aotizhongxin (2013-2017)", loc="center", fontsize=20)
plt.xlabel("Tahun", fontsize=12)
plt.ylabel("Curah Hujan (mm)", fontsize=12)
plt.xticks(df["year"], fontsize=10)
plt.yticks(fontsize=10)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

**Insight:**
- curah hujan tertinggi terjadi pada tahun 2013
- curah hujan terendah terjadi pada tahun 2017
